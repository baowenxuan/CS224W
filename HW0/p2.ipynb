```python
import snap
import numpy as np
import matplotlib.pyplot as plt
```

## 0. Loading Graph

```python
G = snap.LoadEdgeList(snap.PNGraph, "wiki-Vote.txt", 0, 1)
```

## 1. Plot Out-degree Distribution

```python
DegToCntV = snap.TIntPrV()
snap.GetOutDegCnt(G, DegToCntV)
degrees = np.array([p.GetVal1() for p in DegToCntV if p.GetVal1() > 0])
counts = np.array([p.GetVal2() for p in DegToCntV if p.GetVal1() > 0])
```

```python
log10_degrees = np.log10(degrees)
log10_counts = np.log10(counts)
```

```python
plt.plot(log10_degrees, log10_counts, '+-')
plt.show()
```

```python
plt.plot(degrees, counts, '+')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
snap.PlotInDegDistr(G, "Stack-Java", "Stack-Jave In Degree")
```

## 2. Linear Regression

```python
a, b = np.polyfit(x=log10_degrees, y=log10_counts, deg=1)
print(a, b)
```

```python
xx = np.arange(0, 3, 0.1)
yy = a * xx + b
plt.plot(log10_degrees, log10_counts, '+')
plt.plot(xx, yy)
plt.show()
```

```python
xx = np.array([0, 3])
yy = a * xx + b
plt.plot(degrees, counts, '+-')
plt.plot(10**xx, 10**yy)
plt.xscale('log')
plt.yscale('log')
plt.show()
```
